```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
import geopandas as gpd
import datetime

granular_data_18_19 = pd.read_csv('pas_data_ward_level/PAS_ward_level_FY_18_19.csv', low_memory=False)
granular_data_19_20 = pd.read_csv('pas_data_ward_level/PAS_ward_level_FY_19_20.csv', low_memory=False)
```

```python
#Converts the interview date column to an interpretable value

pd.set_option('display.max_columns', None)

def convert_to_standard_date(unique_timestamp):
    unix_timestamp = slope * unique_timestamp + intercept
    return datetime.datetime.utcfromtimestamp(unix_timestamp).strftime('%Y-%m-%d')

slope = 0.991036619270792
intercept = -12095274244.117882

# Apply the conversion function to the 'interview_date' column
granular_data_18_19['converted_date'] = granular_data_18_19['interview_date'].apply(convert_to_standard_date)
granular_data_19_20['converted_date'] = granular_data_19_20['interview_date'].apply(convert_to_standard_date)
```

```python
#revise this list later on
list = ['converted_date','MONTH', 'ward', 'BOROUGHNEIGHBOURHOOD', 'SOA1', 'SOA2' ,'Q62B', 'Q62C', 'Q62F', 'Q62TG', 'NQ135BD', 'RQ80E', 'Q62D', 'Q62E', 'Q62H', 'Q62TI', 'Q62TJ']

filtered_18 = granular_data_18_19[list]
filtered_19 = granular_data_19_20[list]
df2019 = pd.concat([filtered_18, filtered_19], ignore_index=True)

q_conversion = {'Strongly agree': 1, 'Tend to agree': 2, 'Neither agree nor disagree': 3,
'Tend to disagree': 4, 'Strongly disagree': 5, "Don't know": 97,'Refused': 98, 'Not Asked': 99}

q_conversion1 = {'Yes': 1, 'No': 2, "Don't know": 97,'Refused': 98, 'Not Asked': 99}

#Mapping of question answers to values
df2019['Q62B'] = df2019['Q62B'].map(q_conversion)
df2019['Q62C'] = df2019['Q62C'].map(q_conversion)
df2019['Q62F'] = df2019['Q62F'].map(q_conversion)
df2019['Q62TG'] = df2019['Q62TG'].map(q_conversion)
df2019['NQ135BD'] = df2019['NQ135BD'].map(q_conversion)
df2019['Q62D'] = df2019['Q62D'].map(q_conversion)
df2019['Q62E'] = df2019['Q62E'].map(q_conversion)
df2019['Q62H'] = df2019['Q62H'].map(q_conversion)
df2019['Q62TI'] = df2019['Q62TI'].map(q_conversion)
df2019['Q62TJ'] = df2019['Q62TJ'].map(q_conversion)
df2019['RQ80E'] = df2019['RQ80E'].map(q_conversion1)

#interpolates based on column
df2019 = df2019.fillna(df2019.interpolate(axis=0))

#Week conversion
df2019['converted_date'] = pd.to_datetime(df2019['converted_date'])
df2019['week'] = df2019['converted_date'].dt.isocalendar().week
df2019 = df2019[df2019['converted_date'].dt.year == 2019]

#Chaning formatting of Borough names
df2019['BOROUGH'] = df2019['BOROUGHNEIGHBOURHOOD'].str.split(' - ').str[0]
display(df2019)

df2019.to_csv('df2019.csv', index=False)
```

```python
#reading crime data

crimeLSOA = pd.read_csv('MPS LSOA Level Crime (Historical) (1).csv', low_memory=False)
columns_to_keep = [col for col in crimeLSOA.columns if col.startswith('2019')]
non_numeric_columns = [col for col in crimeLSOA.columns if not col.isnumeric()]
final_columns = columns_to_keep + non_numeric_columns
crimeLSOA = crimeLSOA[final_columns]

#display(crimeLSOA)
```

```python
#merged_df = pd.merge(df2019, crime_pivot, left_on=['SOA1', 'week'], right_on=['LSOA Code', 'week'])
#display(crime_pivot)
```

```python
crimeLSOA_melted = pd.melt(crimeLSOA, 
                           id_vars=['LSOA Code', 'LSOA Name', 'Borough', 'Major Category', 'Minor Category'], 
                           var_name='date', 
                           value_name='value')

# Convert the 'date' column to datetime format
crimeLSOA_melted['date'] = pd.to_datetime(crimeLSOA_melted['date'], format='%Y%m')

# Sort by date if needed
crimeLSOA_melted = crimeLSOA_melted.sort_values(by='date')
crimeLSOA_melted['week'] = crimeLSOA_melted['date'].dt.isocalendar().week
    
#crimeLSOA_melted

crime_pivot = crimeLSOA_melted.pivot_table(
    index=["LSOA Code", "LSOA Name", "Borough", "date", "week"], 
    columns="Minor Category", 
    values="value", 
    fill_value=0
).reset_index()

# Display the transformed DataFrame
display(crime_pivot)
```

```python
def duplicate_rows_with_progress(df):
    # Create a list to store duplicated rows
    duplicated_rows = []
    
    total_rows = len(df)
    
    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        # Get the week value of the current row
        week = row['week']
        
        # Duplicate the row for the next 3 weeks
        for i in range(week + 1, week + 4):
            # Create a copy of the current row
            duplicated_row = row.copy()
            
            # Update the week value in the duplicated row
            duplicated_row['week'] = i
            
            # Append the duplicated row to the list
            duplicated_rows.append(duplicated_row)
        
        # Calculate progress
        progress = (index + 1) / total_rows * 100
        print(f"Progress: {progress:.2f}% Complete", end='\r')
        
        # Add the original row itself to the duplicated rows list
        duplicated_rows.append(row)
    
    print("Progress: 100% Complete")
    
    # Create a new DataFrame from the list of duplicated rows
    result_df = pd.DataFrame(duplicated_rows)
    
    return result_df

# Apply the function to your DataFrame
crime_pivot_duplicate = duplicate_rows_with_progress(crime_pivot)
```

```python
cd = crime_pivot_duplicate.sort_values(by=['week', 'LSOA Code'])

cd = crime_pivot_duplicate

def extract_borough_name(row):
    return ' '.join(row['LSOA Name'].split()[:-1])
# Apply the function to create the new column
crime_pivot_duplicate['BoroughName'] = crime_pivot_duplicate.apply(extract_borough_name, axis=1)
# Remove the numeric part from LSOA Name
crime_pivot_duplicate['LSOA Name'] = crime_pivot_duplicate['LSOA Name'].str.replace(r'\d+', '')
cd = crime_pivot_duplicate

display(cd)

#display(cd[cd['LSOA Code'] == 'E01000006'])
```
