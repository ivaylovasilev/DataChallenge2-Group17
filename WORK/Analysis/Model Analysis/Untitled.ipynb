```python
def calculate_percentages(df, group_column, value_columns):
    grouped = df.groupby(group_column)

    for value, group_df in grouped:
        counts = {col: 0 for col in value_columns}
        total_count = 1

        # Sum counts for each value column
        for col in value_columns:
            if col in group_df.columns:
                counts[col] += group_df[col].sum()
                total_count += counts[col]

        # Calculate percentages
        percentages = {col: (counts[col] / total_count) * 100 for col in value_columns}

        # Print the percentages
        print(f"For '{group_column}' is {value}:")
        for col, percentage in percentages.items():
            print(f"{col}: {percentage:.2f}%")
        print()

Ages = ['136r_16-24', 'Q136r_25-34', 'Q136r_35-44', 'Q136r_45-54', 'Q136r_55-64', 'Q136r_65 or over' ]
Employment = ['Q139r_House Person', 'Q139r_Not Working', 'Q139r_Other', 'Q139r_Retired', 'Q139r_Student/Education', 'Q139r_Unemployed', 'Q139r_Working FT', 'Q139r_Working PT']
Ethnicity = ['NQ147r_Asian', 'NQ147r_Black', 'NQ147r_Mixed', 'NQ147r_Other', 'NQ147r_White British', 'NQ147r_White Other']
Religion = ['NQ149r_Christian', 'NQ149r_Hindu', 'NQ149r_Muslim', 'NQ149r_No religion', 'NQ149r_Other religion']
Sexuality = ['Q150r_Heterosexual', 'Q150r_Non-Heterosexual']

# Example usage:
#calculate_percentages(PAS_Q13_encoded, 'Q14C_What you have seen or heard in the media', Sexuality)
calculate_percentages(PAS_NQ79A, 'NQ79A', response_options)

#display(PAS_Q13_encoded)
```
