```python
import pandas as pd
import numpy as np

def interpolate_values(df1, df2, date, year1=2011, year2=2021):
    """
    Interpolates values between two dataframes for a given date.
    """
    if not isinstance(date, pd.Timestamp):
        raise ValueError("The date must be a pandas Timestamp object.")

    common_geographies = set(df1['geography']).intersection(set(df2['geography']))

    df1_filtered = df1[df1['geography'].isin(common_geographies)].copy()
    df2_filtered = df2[df2['geography'].isin(common_geographies)].copy()

    df1_filtered.set_index('geography', inplace=True)
    df2_filtered.set_index('geography', inplace=True)

    df1_filtered = df1_filtered.loc[common_geographies]
    df2_filtered = df2_filtered.loc[common_geographies]

    year_fraction = (date.year + date.day_of_year / 365.25 - year1) / (year2 - year1)

    interpolated_values = df1_filtered + year_fraction * (df2_filtered - df1_filtered)
    interpolated_values = interpolated_values.round(2)
    
    interpolated_values.dropna(axis=1, how='all', inplace=True)

    interpolated_values.sort_values(by='geography', inplace=True)
    interpolated_values.reset_index(inplace=True)
    
    return interpolated_values
```

```python
time = pd.Timestamp('2019-06-15')

cob_int = interpolate_values(dfl_cob11, dfl_cob, time)
yar_int = interpolate_values(dfl_yar11, dfl_yar, time)
eth_int = interpolate_values(dfl_eth11, dfl_eth, time)
eng_int = interpolate_values(dfl_eng11, dfl_eng, time)
rel_int = interpolate_values(dfl_rel11, dfl_rel, time)
acc_int = interpolate_values(dfl_acc11, dfl_acc, time)
rms_int = interpolate_values(dfl_rms11, dfl_rms, time)
whr_int = interpolate_values(dfl_whr11, dfl_whr, time)
occ_int = interpolate_values(dfl_occ11, dfl_occ, time)
eac_int = interpolate_values(dfl_eac11, dfl_eac, time)
edu_int = interpolate_values(dfl_edu11, dfl_edu, time)

acc_int['Total'] = acc_int.loc[:, acc_int.columns != 'geography'].sum(axis=1)
rms_int['Total'] = rms_int.loc[:, rms_int.columns != 'geography'].sum(axis=1)

cob_int.loc[:, cob_int.columns != 'geography'] = cob_int.loc[:, cob_int.columns != 'geography'].div(cob_int['Total'], axis=0)
yar_int.loc[:, yar_int.columns != 'geography'] = yar_int.loc[:, yar_int.columns != 'geography'].div(yar_int['Year of arrival in the UK: Total: All usual residents; measures: Value'], axis=0)
eth_int.loc[:, eth_int.columns != 'geography'] = eth_int.loc[:, eth_int.columns != 'geography'].div(eth_int['Total: All usual residents'], axis=0)
eng_int.loc[:, eng_int.columns != 'geography'] = eng_int.loc[:, eng_int.columns != 'geography'].div(eng_int['Total: All usual residents aged 3 years and over'], axis=0)
rel_int.loc[:, rel_int.columns != 'geography'] = rel_int.loc[:, rel_int.columns != 'geography'].div(rel_int['All usual residents'], axis=0)
acc_int.loc[:, acc_int.columns != 'geography'] = acc_int.loc[:, acc_int.columns != 'geography'].div(acc_int['Total'], axis=0)
rms_int.loc[:, rms_int.columns != 'geography'] = rms_int.loc[:, rms_int.columns != 'geography'].div(rms_int['Total'], axis=0)
whr_int.loc[:, whr_int.columns != 'geography'] = whr_int.loc[:, whr_int.columns != 'geography'].div(whr_int[' Total: All usual residents aged 16 years and over in employment the week before the census'], axis=0)
occ_int.loc[:, occ_int.columns != 'geography'] = occ_int.loc[:, occ_int.columns != 'geography'].div(occ_int['Total: All usual residents aged 16 years and over in employment the week before the census'], axis=0)
eac_int.loc[:, eac_int.columns != 'geography'] = eac_int.loc[:, eac_int.columns != 'geography'].div(eac_int['Total: All usual residents aged 16 years and over'], axis=0)
edu_int.loc[:, edu_int.columns != 'geography'] = edu_int.loc[:, edu_int.columns != 'geography'].div(edu_int['Total: All usual residents aged 16 years and over'], axis=0)
#cob, yar, eth, eng, rel, acc, rms, whr, occ, eac, edu
```

```python
import warnings
import pandas as pd
import numpy as np

warnings.simplefilter(action='ignore', category=FutureWarning)

date_range = pd.date_range(start='2015-01-01', end='2019-12-31')

results = []

def normalize(df, total_column, geography_column='geography'):
    df = df.copy()
    df.loc[:, df.columns != geography_column] = df.loc[:, df.columns != geography_column].div(df[total_column], axis=0)
    return df

for date in date_range:
    time = pd.Timestamp(date)

    cob_int = interpolate_values(dfl_cob11, dfl_cob, time)
    yar_int = interpolate_values(dfl_yar11, dfl_yar, time)
    eth_int = interpolate_values(dfl_eth11, dfl_eth, time)
    eng_int = interpolate_values(dfl_eng11, dfl_eng, time)
    rel_int = interpolate_values(dfl_rel11, dfl_rel, time)
    acc_int = interpolate_values(dfl_acc11, dfl_acc, time)
    rms_int = interpolate_values(dfl_rms11, dfl_rms, time)
    whr_int = interpolate_values(dfl_whr11, dfl_whr, time)
    occ_int = interpolate_values(dfl_occ11, dfl_occ, time)
    eac_int = interpolate_values(dfl_eac11, dfl_eac, time)
    edu_int = interpolate_values(dfl_edu11, dfl_edu, time)

    acc_int['Total'] = acc_int.loc[:, acc_int.columns != 'geography'].sum(axis=1)
    rms_int['Total'] = rms_int.loc[:, rms_int.columns != 'geography'].sum(axis=1)

    cob_int = normalize(cob_int, 'Total')
    yar_int = normalize(yar_int, 'Year of arrival in the UK: Total: All usual residents; measures: Value')
    eth_int = normalize(eth_int, 'Total: All usual residents')
    eng_int = normalize(eng_int, 'Total: All usual residents aged 3 years and over')
    rel_int = normalize(rel_int, 'All usual residents')
    acc_int = normalize(acc_int, 'Total')
    rms_int = normalize(rms_int, 'Total')
    whr_int = normalize(whr_int, ' Total: All usual residents aged 16 years and over in employment the week before the census')
    occ_int = normalize(occ_int, 'Total: All usual residents aged 16 years and over in employment the week before the census')
    eac_int = normalize(eac_int, 'Total: All usual residents aged 16 years and over')
    edu_int = normalize(edu_int, 'Total: All usual residents aged 16 years and over')

    merged_df = cob_int.merge(yar_int, on='geography').merge(eth_int, on='geography').merge(eng_int, on='geography').merge(rel_int, on='geography').merge(acc_int, on='geography').merge(rms_int, on='geography').merge(whr_int, on='geography').merge(occ_int, on='geography').merge(eac_int, on='geography').merge(edu_int, on='geography')

    merged_df['date'] = time

    for borough in merged_df['geography']:
        merged_df[f'BOROUGH_{borough}'] = merged_df['geography'] == borough

    results.append(merged_df)

final_df = pd.concat(results, ignore_index=True)

columns_to_remove = final_df.columns[(final_df == 1.0).all()]
final_df = final_df.drop(columns=columns_to_remove)
```
