```python
#The purpose of this file is to:
# take the Youtube and News data
# merge them, and save them into a new file, for later use

import pandas as pd
from pathlib import Path
from sklearn.preprocessing import MinMaxScaler

NEWS1 = pd.read_csv('Data/News/Islington_weekly.csv')
YOUTUBE1 = pd.read_csv('Data/YouTube.csv')
f1519 = pd.read_csv('15-19_data_weekly.csv')
```

```python
pd.set_option('display.max_columns', None)

columns = ["week/0", "week/1", "avg_sentiment", "weighted_sentiment",
            "category", "num_comments", "views", "likes", "title_sentiment", "description_sentiment"]

YOUTUBE = YOUTUBE1[columns]
YOUTUBE['category'] = YOUTUBE['category'].fillna('YTUnidentified_Category')
YOUTUBE = pd.get_dummies(YOUTUBE, columns=['category'], prefix='YT', prefix_sep='')

YOUTUBE['week'] = pd.to_datetime(YOUTUBE['week/1'].astype(str) + '-W' + YOUTUBE['week/0'].astype(str) + '-1', format='%G-W%V-%u')
cols = ['week'] + [col for col in YOUTUBE.columns if col != 'week']
YOUTUBE = YOUTUBE[cols]
YOUTUBE = YOUTUBE.drop(['week/0', 'week/1'], axis = 1 )

YOUTUBE['likes'] = YOUTUBE['likes']/YOUTUBE['views']
YOUTUBE['avg_sentiment'] = YOUTUBE['avg_sentiment']*YOUTUBE['views']*0.01

YOUTUBE_weekly = YOUTUBE.drop(['likes', 'views'], axis=1)

YOUTUBE = YOUTUBE.astype({col: int for col in YOUTUBE.select_dtypes(include='bool').columns})

week_columns = ["week"]
YOUTUBE_weekly = YOUTUBE.groupby(week_columns).mean().reset_index()


scaler = MinMaxScaler(feature_range=(-1, 1))
columns_to_normalize = YOUTUBE_weekly.columns.difference(['week', 'YTYTUnidentified_Category', 'YTcommunity relations', 'YTdrugs', 'YTfirearm',
                                                          'YTillegal immigration', 'YTlegal and policy', 'YTpolice accountability', 'YTprotest',
                                                          'YTterrorism', 'YTtheft', 'YTviolence'])
YOUTUBE_weekly[columns_to_normalize] = scaler.fit_transform(YOUTUBE_weekly[columns_to_normalize])

YOUTUBE_weekly['week'] = pd.to_datetime(YOUTUBE_weekly['week'])  # Convert 'week' to datetime if not already

#YOUTUBE = pd.get_dummies(YOUTUBE1, columns=['borough'], prefix='YT', prefix_sep='')
# YOUTUBE = YOUTUBE.rename(columns={'YTUnknown': 'YTAllLondon'})
#??CONSIDERING REMOVING VIDEOS WITH NO COMMENTS??
#display(YOUTUBE.loc[YOUTUBE['weighted_sentiment'] == 0])

# YOUTUBE2 = YOUTUBE[['week/0', 'week/1', 'num_mentions', 'avg_sentiment', 'weighted_sentiment',  'total_weights', 'num_comments', 'views',
#                     'likes', 'title_sentiment', 'description_sentiment', 'YTBarking and Dagenham', 'YTBarnet', 'YTBexley', 'YTBrent', 'YTCamden', 
#                     'YTEaling', 'YTEnfield', 'YTGreenwich', 'YTHackney', 'YTHaringey', 'YTHarrow', 'YTHavering', 'YTIslington', 
#                     'YTKensington and Chelsea',  'YTKingston upon Thames', 'YTLambeth', 'YTLewisham', 'YTMerton', 'YTRedbridge',
#                     'YTRichmond upon Thames', 'YTTower Hamlets', 'YTAllLondon','YTWaltham Forest', 'YTWestminster', 'YTUnidentified_Category', 
#                     'community relations', 'drugs', 'firearm', 'illegal immigration', 'legal and policy', 'police accountability', 'protest',
#                     'terrorism', 'theft', 'violence']]

# YOUTUBE2 = YOUTUBE[['borough']]

# #display(YOUTUBE2[YOUTUBE2['week'] == '2018-01-08'] )
# #display(YOUTUBE2[YOUTUBE2['week'].duplicated(keep=False)])


# numeric_means = YOUTUBE2.groupby('week').mean()
# one_hot_means = YOUTUBE2.groupby('week').mean()
# YOUTUBE3 = pd.concat([one_hot_means], axis=1)
# YOUTUBE3.reset_index(inplace=True)

display(YOUTUBE_weekly)
```

```python
#combining the individual borough level dataframes 

directory_path = Path('Data/News')

mylist = []
count = 0
for file in directory_path.iterdir():
    if file.is_file():
        a = pd.read_csv(file)
        name = file.name
        name_store = name
        name = a.copy()
        name.columns = [f"{col}_{name.loc[0, col]}" for col in name.columns]
        name = name.drop(index = [0,1])
        name.columns = [f"{name_store}_{col}" for col in name.columns]
        name.columns.values[0] = 'week'
        mylist.append(name)
        count+= 1
NEWS = pd.concat(mylist, axis= 0)
print(count)
NEWS = NEWS.groupby('week', as_index=False).first()
NEWS.fillna(0, inplace=True)

display(NEWS)
```

```python
first_row = f1519.iloc[0]

for column_name in f1519.columns:
    f1519.rename(columns={column_name: column_name + "_" + str(first_row[column_name])}, inplace=True)

f1519 = f1519.iloc[2:]
column_to_exclude = 'Unnamed: 0_Category'
columns_to_convert = [col for col in f1519.columns if col != column_to_exclude]
f1519[columns_to_convert] = f1519[columns_to_convert].astype(float)

f1519['avg_sentiment_drugs'] = f1519['avg_sentiment_drugs'] * f1519[ 'num_articles_drugs'] * f1519['num_mentions_drugs']
f1519['avg_sentiment.1_firearm'] = f1519['avg_sentiment.1_firearm'] * f1519[ 'num_articles.1_firearm'] * f1519['num_mentions.1_firearm']
f1519['avg_sentiment.2_protest'] = f1519['avg_sentiment.2_protest'] * f1519[ 'num_articles.2_protest'] * f1519['num_mentions.2_protest']
f1519['avg_sentiment.3_terrorism'] = f1519['avg_sentiment.3_terrorism'] * f1519[ 'num_articles.3_terrorism'] * f1519['num_mentions.3_terrorism']
f1519['avg_sentiment.4_theft'] = f1519['avg_sentiment.4_theft'] * f1519[ 'num_articles.4_theft'] * f1519['num_mentions.4_theft']
f1519['avg_sentiment.5_violence'] = f1519['avg_sentiment.5_violence'] * f1519[ 'num_articles.5_violence'] * f1519['num_mentions.5_violence']

f1519 = f1519[['Unnamed: 0_Category', 'avg_sentiment_drugs',	'avg_sentiment.1_firearm',	'avg_sentiment.2_protest',	'avg_sentiment.3_terrorism',
            'avg_sentiment.4_theft', 'avg_sentiment.5_violence']]

display(f1519)
```
